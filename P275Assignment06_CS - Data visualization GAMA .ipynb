# How many ways to convey information in an XY plot?

* X-position
* Y-position
* Color
* Transparency
* Size
* Marker
* Left y-axis
* Top x-axis

```python
#Do imports

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
from matplotlib import rcParams
rcParams["savefig.dpi"] = 150
rcParams["figure.dpi"] = 150
rcParams["font.size"] = 15
```

```python

x = np.linspace(-3,3,100)
y = (x**2) + np.random.randn(100) # what is this doing? rand vs randn
colors = np.random.uniform(size=100) #same
sizes = 100 * np.random.uniform(size=100) 

x2 = x/np.pi
y2 = np.pi*y

# oh and we want to compare to the model y = x^2


fig = plt.figure(figsize=(8,4))

ax1 = plt.axes()
c = ax1.scatter(x, y, c=colors, s=100*colors, alpha=0.4,
            cmap='viridis',label='data') #viridis or bwr or tab20 or gist_gray
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax1.plot(x,x**2,'k--',alpha=0.4,linewidth=1,label='model')

# second axis
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
y2 = -np.cos(x)
y2 = np.pi*y
ax2.plot(x,y2,'k,',alpha=0)
ax2.set_ylabel('Y2')

# third axis
ax3 = ax2.twiny()  # instantiate a second axes that shares the same x-axis
ax3.plot(x2,y,'k,',alpha=0)
ax3.set_xlabel('X2');

fig.colorbar(c, label='colorscale',pad=0.12); 
ax1.legend(loc='center',fontsize=9)

```

# Load in the data from GAMA data

The file GAMA.csv contains information on galaxies from the Galaxy And Mass Survey. 
Each column is a different characteristic of each GAMA Galaxy.

0 - GAMA CATAID (x-axis)

1 - Stellar mass (log10 solar masses) (y-axis1)

2 - u-r colour (color of points)

3 - S'ersic index (log10) (marker)

4 - Half-light radius (log10 kpc) (x-axis2)

5 - Specific star formation rate (log10 Gyr^-1) (y-axis2)

Stellar mass is the order magnitude of the total mass of stars. u-r is the color in magnitudes with higher values meaning redder and lower values bluer. The Sersic index is the kind of profile, an index of 1 or 2 is a disk galaxy, an index of 4 is a spheroid galaxy. The half-light radius is the typical size of the galaxy. The last column is how fast it is making new stars.
